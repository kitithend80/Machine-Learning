```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("student_data.csv")

print("First five records:")
print(df.head())
print("--------------------------------------------------------------------------------------")
print("Last five records:")
print(df.tail())
print("--------------------------------------------------------------------------------------")
print("Total numbr of rows and columns:")
print(df.shape)
print("--------------------------------------------------------------------------------------")
print("Names of columns:")
print(df.dtypes)
print("--------------------------------------------------------------------------------------")
print("Data set infromation:")
print(df.info)
print("--------------------------------------------------------------------------------------")
```

```python
mean=df['Attendance'].mean()
print("Mean Attendance:",mean)
print("--------------------------------------------------------------------------------------")
print("Average Internal Marks:",df['InternalMArks'].mean())
print("--------------------------------------------------------------------------------------")
print("Average External Marks:",df['ExternalMarks'].mean())
print("--------------------------------------------------------------------------------------")
print("Maximum Internal MArks:",df['InternalMArks'].max())
print("--------------------------------------------------------------------------------------")
print("Minimum Internal Marks:",df['InternalMArks'].min())
print("--------------------------------------------------------------------------------------")
print("Maximum External MArks:",df['ExternalMarks'].max())
print("--------------------------------------------------------------------------------------")
print("Minimum External MArks:",df['ExternalMarks'].min())
print("--------------------------------------------------------------------------------------")
print("Standard deviation of Internal marks:",df['InternalMArks'].std())
print("--------------------------------------------------------------------------------------")
print("Summary Statistics:")
print(df.describe())
```

```python
df['TotalMarks']=df['InternalMArks']+df['ExternalMarks']
df['Percentage']=(df['TotalMarks']/100)*100
print("Updated Dataframe:")
print(df)
```

```python
print("Students attendance greater than 90%:")
print(df[df['Attendance']>90])
print("--------------------------------------------------------------------------------------")
print("students with total marks greater than 90:")
print(df[df['TotalMarks']>90])
print("--------------------------------------------------------------------------------------")
print("Female students:")
print(df[df['Gender']=='F'])
print("--------------------------------------------------------------------------------------")
print("male students:")
print(df[df['Gender']=='M'])
print("--------------------------------------------------------------------------------------")
print("Students with external marks less than 50:")
print(df[df['ExternalMarks']<50])
```

```python
highest=df.loc[df['TotalMarks'].idxmax()]
print(highest)
print("---------------------------------------------")
lowest=df.loc[df['TotalMarks'].idxmin()]
print(lowest)
print("---------------------------------------------")
print("top three:",df.nlargest(3,'TotalMarks'))
print("---------------------------------------------")
print("bottom three:",df.nsmallest(3,'TotalMarks'))
print("---------------------------------------------")
above_avg=len(df[df['TotalMarks']>mean])
print("no of students above avg are:",above_avg)
```

```python
print("Average Total marks of male students:")
print(df[df['Gender']=='M']['TotalMarks'].mean())
print("Average Total marks of female students:")
print(df[df['Gender']=='F']['TotalMarks'].mean())
print(df.groupby('Gender')['Attendance'].mean())
```
